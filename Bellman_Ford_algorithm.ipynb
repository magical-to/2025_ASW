```python
%%writefile main.cpp
#include <iostream>
#include <vector>
#include <fstream>
#include <limits>

struct Edge {
    int u;
    int v;
    int w;
};

const long long INF = std::numeric_limits<long long>::max();

int main() {
    std::ifstream inputFile("input.txt");
    if (!inputFile.is_open()) {
        std::cerr << "input.txt 읎어요" << std::endl;
        return 1;
    }

    int V, E;
    inputFile >> V >> E;

    std::vector<Edge> edges;
    for (int i = 0; i < E; ++i) {
        int u, v, w;
        inputFile >> u >> v >> w;
        edges.push_back({u, v, w});
    }
    inputFile.close();

    int startNode = 1;

    std::vector<long long> dist(V + 1, INF);

    dist[startNode] = 0;

    for (int i = 1; i < V; ++i) {
        for (const auto& edge : edges) {
            if (dist[edge.u] != INF && dist[edge.u] + edge.w < dist[edge.v]) {
                dist[edge.v] = dist[edge.u] + edge.w;
            }
        }
    }

    bool negativeCycle = false;
    for (const auto& edge : edges) {
        if (dist[edge.u] != INF && dist[edge.u] + edge.w < dist[edge.v]) {
            negativeCycle = true;
            break;
        }
    }

    if (negativeCycle) {
        std::cout << "Negative cycle exists" << std::endl;
    } else {
        for (int i = 1; i <= V; ++i) {
            std::cout << i << " --> " << dist[i] << std::endl;
        }
    }

    return 0;
}
```
